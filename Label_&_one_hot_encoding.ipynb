```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data={"color":["red","blue","green","yellow","purple"]}
df=pd.DataFrame(data)
df.head()
```

```python
lb=LabelEncoder()
df["color_label"]=lb.fit_transform(df["color"])
```

```python
df
```

```python
one_hot_encoder=OneHotEncoder(sparse_output=False)
one_hot_encoded_data=one_hot_encoder.fit_transform(df[["color"]])
```

```python
one_hot_encoded_data
```

```python
color_encoded_df=pd.DataFrame(one_hot_encoded_data,columns=one_hot_encoder.get_feature_names_out(["color"]))
```
